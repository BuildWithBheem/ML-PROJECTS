```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Placement_Data.csv')
```

```python
df = df.drop(columns = ['salary','sl_no','workex','etest_p','specialisation','mba_p']) # Not neccessary for placements in college
```

```python
print(df.columns)
```

```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
ct  = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[0,2,4,5,7])],remainder = 'passthrough') #Encode Categorical columns
x = ct.fit_transform(x)
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(ct.feature_names_in_)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()               # Feature Scaling
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'poly', degree = 3)

svc.fit(x_train,y_train)

x_predict3 = svc.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score
cm  = confusion_matrix(y_test,x_predict3)
print(cm)
print(accuracy_score(y_test,x_predict3))
```

```python
import joblib as jb     # Export Model as a file
jb.dump(ct, "encoder_new")
jb.dump(sc, "scaler_new")
jb.dump(svc, "model_new")
```
